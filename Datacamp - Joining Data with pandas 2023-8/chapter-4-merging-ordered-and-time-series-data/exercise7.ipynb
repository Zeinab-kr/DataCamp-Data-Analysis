````python
from pickle import load
import urllib.request
import pandas as pd

social_fin = pd.read_pickle("social_financials.p")
````

````python
social_fin
````

````python
social_fin.query("value > 50000000").shape
````

````python
social_fin.query('company == "Facebook"').shape
````

````python
social_fin.query('(financial == "net_income") and (value < 0)').shape
````

````python
social_fin.query('(financial == "gross_profit") and (value > 100)').shape
````
