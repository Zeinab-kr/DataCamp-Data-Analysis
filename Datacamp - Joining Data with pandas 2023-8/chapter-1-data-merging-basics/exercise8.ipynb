````python
from pickle import load
import urllib.request
import pandas as pd
````

````python
fn_lic = 'https://assets.datacamp.com/production/repositories/5486/datasets/2a4d8e5d91f6f2b41477fa6ea81da91e4f09305e/licenses.p'
fn_land = 'https://assets.datacamp.com/production/repositories/5486/datasets/d9c181b64947bf60ddd9770bad424db3717272e9/land_use.p'
fn_cen = 'https://assets.datacamp.com/production/repositories/5486/datasets/c778c41d8b2f0d4cc6c842e95e0c650c7156c82c/census.p'
licenses = load(urllib.request.urlopen(fn_lic))
land_use = load(urllib.request.urlopen(fn_land))
census = load(urllib.request.urlopen(fn_cen))
````


- Merge `land_use` and `census` on the `ward` column. Merge the result of this with `licenses` on the `ward` column, using the suffix `_cen` for the left table and `_lic` for the right table. Save this to the variable `land_cen_lic`.

````python
land_cen_lic = land_use.merge(census, on="ward").merge(licenses, on="ward", suffixes=["_cen", "_lic"])
````

````python
land_cen_lic.head()
````

- Group `land_cen_lic` by  `ward`, `pop_2010` (the population in 2010), and `vacant`, then count the number of `accounts`. Save the results to `pop_vac_lic`.

````python
pop_vac_lic = land_cen_lic.groupby(["ward", "pop_2010", "vacant"]).agg({'account':'count'})
````

````python
pop_vac_lic.head()
````

- Sort `pop_vac_lic` by `vacant`, `account`, and`pop_2010` in descending, ascending, and ascending order respectively. Save it as `sorted_pop_vac_lic`.

````python
pop_vac_lic.sort_values(["vacant", "account", "pop_2010"], ascending=[False, True, True]).head(10)
````
