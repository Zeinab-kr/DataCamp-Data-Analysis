##  Assignment

To do this, we'll need to use a count plot. Count plots take in a categorical list and return bars that represent the number of list entries per category. You can create one here using a list of regions for each country, which is a variable named `region`.

````python
import pandas as pd

# This dataset represents decimals with a comma so we need to convert it back
country_data = pd.read_csv('https://assets.datacamp.com/production/repositories/3996/datasets/cd320cdd281edc2d2ce24058565d3d7090aa9708/countries-of-the-world.csv', decimal=",")

# Set up variable for scatter plot (& remove trailing spaces)
region = country_data["Region"]
region = list(region.apply(lambda x: x.strip()))

# Clean up
del country_data
````

##  Instructions

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
sns.countplot(region)
````
